```python
import pandas as pd

CSV_PATH = './pg_catalog.csv'

csv = pd.read_csv(CSV_PATH, index_col=0)

# get unique subjects
subjects = csv['Subjects'].unique()
for subject in subjects:
    if 'math' in str(subject).lower():
        print(subject)

# get unique bookshelves
shelves = csv['Bookshelves'].unique()
for shelf in shelves:
    if 'math' in str(shelf).lower():
        print(shelf)
```

```python
# find all with math in subjects or bookshelves
math_csv = csv[csv['Subjects'].str.contains('math', case=False) | csv['Bookshelves'].str.contains('math', case=False)]
```
